## String Formatting

Python uses C-style string formatting to create new, formatted strings. The "%" operator is used to format a set of variables enclosed in a "tuple" (a fixed size list), together with a format string, which contains normal text together with "argument specifiers", special symbols like "%s" and "%d".

    %s - String (or any object with a string representation, like numbers)

    %d - Integers

    %f - Floating point numbers

    %.<number of digits>f - Floating point numbers with a fixed amount of digits to the right of the dot.

    %x/%X - Integers in hex representation (lowercase/uppercase)

```python
name = "John"
age = 23
print("%s is %d years old." % (name, age))

mylist = [1,2,3]
print("A list: %s" % mylist)
```

```python
astring = "hello rahul's world."
print(astring, " " , len(astring)) 
#calculate the lenght of string

print(astring.index("r")) 
# find the position of givven element

print(astring.count("z")) 
# count the element "z" present in the string
print(astring.count("l"))

print(astring[0:9:2]) 
# string slicing

print(astring.upper()) 
# convert string into uppercase

print(astring.lower()) 
# convert string into lowercase

print(astring.startswith("hello")) 
# check whether string start with given word or not (return true and false)

print(astring.endswith("asdfasdfasdf")) 
# check whether string end with given word or not (return true and false)

afewwords = astring.split(" ") 
# split the string and form a list 

print(afewwords)
```

```python
s = "Strings are awesome!"
# Length should be 20
print("Length of s = %d" % len(s))

# First occurrence of "a" should be at index 8
print("The first occurrence of the letter a = %d" % s.index("a"))

# Number of a's should be 2
print("a occurs %d times" % s.count("a"))

# Slicing the string into bits
print("The first five characters are '%s'" % s[:5]) # Start to 5
print("The next five characters are '%s'" % s[5:10]) # 5 to 10
print("The thirteenth character is '%s'" % s[12]) # Just number 12
print("The characters with odd index are '%s'" %s[1::2]) #(0-based indexing)
print("The last five characters are '%s'" % s[-5:]) # 5th-from-last to end

# Convert everything to uppercase
print("String in uppercase: %s" % s.upper())

# Convert everything to lowercase
print("String in lowercase: %s" % s.lower())

# Check how a string starts
if s.startswith("Str"):
    print("String starts with 'Str'. Good!")

# Check how a string ends
if s.endswith("ome!"):
    print("String ends with 'ome!'. Good!")

# Split the string into three separate strings,
# each containing only a word
print("Split the words of the string: %s" % s.split(" "))
```

## Dictionary

Python Dictionary is used to store the data in a key-value pair format. The dictionary is the data type in Python, which can simulate the real-life data arrangement where some specific value exists for some particular key. It is the mutable data-structure. The dictionary is defined into element Keys and values.

- Keys must be a single element
- Value can be any type such as list, tuple, integer, etc.

```python
mydic = {"name" : "rahul" , "age" : 23 , "salary" : "very little" ,
         "interest" : "data science", "cgpa" : 7.51}

print(type(mydic))
print(mydic.values())
print(mydic.keys())

print(mydic)
print(mydic["name"])
del mydic["name"]          # to delete the name key , we use "del"
mydic.pop("age")  # we can also use pop method to remove a element , pop method returns a value whereas del doesn't return any value
print(mydic)
mydic["hello"] = "world" , 1 , 2

print(mydic)
```

## Date and Time

The datetime classes are classified in the six main classes.

    date - It is a naive ideal date. It consists of the year, month, and day as attributes.
-----------------------------------------------------------
    time - It is a perfect time, assuming every day has precisely 24*60*60 seconds. It has hour, minute, 
    second, microsecond, and tzinfo as attributes.
-----------------------------------------------------------
    datetime - It is a grouping of date and time, along with the attributes year, month, day, hour, minute, second, microsecond, and tzinfo.
-----------------------------------------------------------
    timedelta - It represents the difference between two dates, time or datetime instances to microsecond resolution.
-----------------------------------------------------------
    tzinfo - It provides time zone information objects.
-----------------------------------------------------------
    timezone - It is included in the new version of Python. It is the class that implements the tzinfo abstract base class.

```python
import time

print(time.localtime(time.time()))  
```

```python
print(time.asctime(time.localtime(time.time())))
```
